# 1. Data loading

```python
ids, predictions, data = load_data("train.csv")
perc = np.count_nonzero(predictions)/len(predictions)
print("The percentage of Boson events in this set is: ", perc*100)
```

```python
""" Here we try to clean the data
# we could do something like: if a column is made of more then 90% of -999, we eliminate that column, and hence condense the array "data". The same
# could be done with rows (data points). Then, we can run other few lines of code to substitute the remaining -999 with the mean value for that feature
# in this way we should obtain usable data


# initially, we have 29 columns
num = 0.7        # this is the parameter through which we decide to eliminate a column or not. with 0.7 we lose 7 columns
length = data.shape[1]  
i = 0
while True:
    count = data[:,i][data[:,i]==-999].shape[0] 
    tot = num*len(data[:,i])        
    if count >= tot:
        data = np.delete(data, i, 1)      # the 1 stands for column
    else:
        i = i+1
    
    if i >=  data.shape[1]:
        break

        
# Now we will try to substitute the remaining -999 values with the average value in that column    
for i in range (0, data.shape[1]):
    
    # Everytime we get to a new column, we set count and sum to 0
    count = 0;
    Sum = 0;
    
    for j in range(0, data.shape[0]):
        if (data[j][i] == -999):       
            continue
        else:
            Sum += data[j][i]
            count += 1
            
    average = Sum / count
    
    # Now we insert the average value in every position where a -999 is located
    for j in range(0, data.shape[0]):
        if (data[j][i] == -999):
            data[j][i] = average
        

# USEFUL PRINTS:
# print(data.shape[0])
# print(data.shape[1])
# print(pred)
# data.shape[0]

# Let's quickly check whether there are any -999 left in the data matrix
#count = 0
#for i in range (0, data.shape[1]):
#    for j in range (0, data.shape[0]):
#        if (data[j][i] == -999):
#            count += 1
            
#print(count)
#print(data)"""
```

## 4.1 Least squares with gradient descent

```python
# We import all the functions
from implementations import*

# We need to add the shift term in the data matrix
# new_column_tr  = np.ones(train.shape[0], )
# new_column_te  = np.ones(test.shape[0], )
# train = np.insert(train, 0, new_column_tr, axis = 1)
# test = np.insert(test, 0, new_column_te, axis = 1)

# The 'y' vector we will use later is the variable 'pred' (numpy array)

# Our data matrix is N * D
D = train.shape[1]

# We will consider as initial guess a vector of zeros
initial_w = np.ones(D, )

# Other parameters
max_iters = 100
gamma     = 0.7

# We compute the optimal weights and the MSE loss via the gradient descent algorithm
[w_gd, loss_gd] = mean_squared_error_gd(preds_train, train, initial_w, max_iters, gamma)

# Now we calculate the loss on the test set
loss_gd_te = compute_MSE(preds_test, test, w_gd)

print(w_gd)
print(loss_gd_te)
```

## 4.4 Ridge regression with normal equations

```python
max_iters = 1000
initial_w = np.zeros(processed_data.shape[1])
lambdas = np.logspace(-4, 0, 30)

k_indices = build_k_indices(preds_train, k_fold, seed)
loss_tr_ridge = []
loss_te_ridge = []
# flag = 0
# flag2 = 0
for lambda_ in lambdas:
    loss_tr_tmp = []
    loss_te_tmp = []
    for k in range(k_fold):
        test_predictions, test_data, training_predictions, training_data = cross_validation(preds_train,train,k_indices,k)
        w_opt = ridge_regression(training_predictions, training_data, lambda_)
        training_loss = compute_MSE(training_predictions, training_data, w_opt)
        
        test_loss = compute_MSE(test_predictions, test_data, w_opt)
        #flag2 = flag2+1
        #if training_loss > test_loss:
         #   flag += 1
        
        loss_tr_tmp.append(training_loss)
        loss_te_tmp.append(test_loss)
    loss_tr_ridge.append(np.mean(loss_tr_tmp))
    loss_te_ridge.append(np.mean(loss_te_tmp))
lambda_opt_ridge = lambdas[np.argmin(loss_te_ridge)]

# After the cross validation, which has led to lambda_opt_ridge,
# we compute the optimal weights and the consequent loss using the train set splitted in 2 

w_opt_ridge = ridge_regression(preds_train, train, lambda_opt_ridge)
test_loss_ridge = compute_MSE(preds_test, test, w_opt_ridge)
```

```python
plt.semilogx(lambdas, loss_tr_ridge, marker=".", color='b', label='train error')
plt.semilogx(lambdas, loss_te_ridge, marker=".", color='r', label='test error')
plt.xlabel("$\lambda$")
plt.ylabel("MSE loss (train and test)")
plt.xlim(1e-10, 1)
plt.title("Cross validation: ridge regression with normal equations")
plt.legend(loc=2)
plt.grid(True)
```

## 4.6 Regularized logistic regression with gradient descent

```python
max_iters = 1000
initial_w = np.zeros(processed_data.shape[1])
gamma = 0.01
lambdas = np.logspace(-7, -2, 30)

k_indices = build_k_indices(preds_train, k_fold, seed)
loss_tr = []
loss_te = []
for lambda_ in lambdas:
    loss_tr_tmp = []
    loss_te_tmp = []
    for k in range(k_fold):
        test_predictions, test_data, training_predictions, training_data = cross_validation(preds_train,train,k_indices,k)
        train_loss, w_opt = reg_logistic_regression(training_predictions, training_data, lambda_, initial_w, max_iters, gamma)
        test_loss = compute_ridge_log_loss(test_predictions, test_data, w_opt)
        loss_tr_tmp.append(train_loss)
        loss_te_tmp.append(test_loss)
    loss_tr.append(np.mean(loss_tr_tmp))
    loss_te.append(np.mean(loss_te_tmp))
lambda_opt = lambdas[np.argmin(loss_te)]
train_loss, w_opt = reg_logistic_regression(training_predictions, training_data, lambda_opt, initial_w, max_iters, gamma)
test_loss = compute_ridge_log_loss(test_predictions, test_data, w_opt)
```

```python
plt.semilogx(lambdas, loss_tr, marker=".", color='b', label='train error')
plt.semilogx(lambdas, loss_te, marker=".", color='r', label='test error')
plt.xlabel("lambda")
plt.ylabel("loss")
plt.xlim(1e-4, 1)
plt.title("cross validation")
plt.legend(loc=2)
plt.grid(True)
```
